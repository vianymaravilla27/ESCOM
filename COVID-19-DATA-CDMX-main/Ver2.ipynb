# POSITIVOS CDMX

```python
import pandas as pd 
import matplotlib.pyplot as plt
from pandas import DataFrame


tasa = pd.read_csv("positivos.csv", usecols= ['tasa_positividad'])
tasa_cdmx = pd.read_csv("positivos.csv", usecols = ['tasa_positividad_cdmx'])

Data= pd.read_csv("positivos.csv", usecols =['fecha_toma_muestra','positivos_totales_cdmx','positivos_totales'] )



df = DataFrame(Data,columns=['fecha_toma_muestra','positivos_totales','positivos_totales_cdmx'])
df.plot(x = 'fecha_toma_muestra', y= 'positivos_totales_cdmx')

plt.show()
```
